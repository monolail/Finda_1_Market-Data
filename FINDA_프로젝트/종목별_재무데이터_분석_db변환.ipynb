```python
import pandas as pd
import sqlite3

df = pd.read_excel(
    "종목별 재무재표 분석.xlsx",
    dtype={"종목코드": str}
)

df["종목코드"] = df["종목코드"].str.zfill(6)
```

```python
companies = (
    df[["종목코드", "기업명"]]
    .drop_duplicates()
    .rename(columns={
        "종목코드": "stock_code",
        "기업명": "corp_name"
    })
)
```

```python
# 컬럼명을 영어로 전환
financials = df.drop(columns=["기업명"]).rename(columns={
    "종목코드": "stock_code",
    "분석연도": "year",

    "ROA": "roa",
    "ROE": "roe",
    "순이익률": "net_margin",

    "매출액증가율" : "sales_growth_rate",
    "매출증가율": "sales_growth",
    "자산증가율": "asset_growth",
    "영업이익증가율": "op_income_growth",
    "순이익증가율": "net_income_growth",
    "영업이익률": "op_margin",

    "부채비율": "debt_ratio",
    "자기자본비율": "equity_ratio",
    "유동비율": "current_ratio",
    "총자산회전율": "asset_turnover_ratio",

    "총자산": "total_assets",
    "유동자산": "current_assets",
    "비유동자산": "non_current_assets",

    "총부채": "total_liabilities",
    "유동부채": "current_liabilities",
    "비유동부채": "non_current_liabilities",

    "자본금": "capital",
    "자본총계": "equity",
    "이익잉여금": "retained_earnings"
})
```

```python
conn = sqlite3.connect("financials_EN.db")

companies.to_sql(
    "companies",
    conn,
    if_exists="replace",
    index=False
)

financials.to_sql(
    "financials",
    conn,
    if_exists="replace",
    index=False
)

conn.close()
```
