## 통계학이란 무엇인가?

통계학은 ‘불완전한 데이터로부터 세상의 구조를 추론하는 학문’이다.
우리는 항상 전체를 볼 수 없고, 일부 관측값(표본)만을 가지고

- 무엇이 일반적인지
- 무엇이 이상한지
- 얼마나 확신할 수 있는지

를 판단해야 한다.

### 일상 예시

- 전 국민의 키를 잴 수 없으니 표본으로 평균 키를 추정
- 동전 10번 던져 앞면 7번 → “이 동전은 공정한가?”
- A 광고 클릭률 3.1%, B 광고 3.4% → “정말 B가 더 좋은가?”

통계학은 계산이 아니라 ‘판단의 학문’

📌 과제 제목

“표본으로부터 전체를 추론하라”

📌 학습 목표

데이터 요약(기술통계)의 의미 이해

표본 평균의 변동성 체감

시각화를 통해 분포 개념 이해

통계적 사고와 단순 계산의 차이 구분

설명

학생들의 키 데이터가 있다고 가정하자.
이 데이터의 중심과 흩어짐을 요약하라.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 가상의 학생 키 데이터 (cm)
heights = np.array([
    168, 172, 170, 165, 180, 175, 178, 169, 171, 174,
    160, 162, 158, 177, 173, 176, 168, 170, 169, 172
])

# 기본 통계량
df = pd.DataFrame(heights, columns=["height"])
df.describe()
```

### 질문

- 평균과 중앙값은 각각 얼마인가?
- 평균과 중앙값이 다른 이유는 무엇일 수 있는가?
- 표준편차는 무엇을 의미하는가?

### 문제 1 : 답변

- 평균과 중앙값은 각각 얼마인가?
  - 평균 : 170.350000 / 중앙값 : 170.500000
- 평균과 중앙값이 다른 이유는 무엇일 수 있는가?
  - 중앙값은 표본에서 데이터를 정렬했을 때의 '중앙에 위치한 값'을 의미하고, 평균은 '표본 데이터 전체 값의 합에 데이터의 갯수'를 나눈 값입니다.
  - 따라서 데이터 분포의 비대칭성과 극단값(outlier)의 영향에 따라, 중앙값과 평균은 충분히 다르게 나타날 수 있습니다.
  - 예를 들면, 160cm이 2명, 180cm이 7명, 170cm이 1명인 데이터에서의 중앙값은 180cm가 되겠지만, 평균은 175cm로 둘 간의 차이가 크다는 것을 알 수 있습니다.
- 표준편차는 무엇을 의미하는가?
  - 표준편차(Standard Deviation)는 데이터가 평균을 중심으로 얼마나 흩어져 있는지(변동성, 산포도)를 알아볼 수 있는 지표입니다. 즉, 표준편차가 크면 데이터가 그만큼 평균에서 멀리 떨어졌서 분포한다는 것을 알 수 있고, 반대로 표준편차가 작다면 그만큼 평균에 가깝다는 얘기가 됩니다.

## 🔹 문제 2. 시각화로 분포 이해하기

히스토그램을 통해 데이터의 형태를 관찰하라.

```python
plt.hist(heights, bins=6)
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Heights")
plt.show()
```

### 질문
- 이 분포는 대칭적인가?
- 극단값(outlier)이 존재하는가?

### 문제 2 : 답변

- 이 분포는 대칭적인가?
  - 그렇지 않습니다.
  - 그림에서 보이듯이, 표의 모양이 우측으로 치우쳐진 그래프임을 알 수 있습니다.
- 극단값(outlier)이 존재하는가?
  - 존재하지 않습니다.
  - 그래프의 모양새는 다소 우측으로 치우쳐진 것으로 보일지라도, 평균과 중앙값이 1분위수(25%)와 3분위수(75%)사이에 값이 있는 것으로 보아 극단값이 있다고 보기 어렵습니다.
  - 따라서 사분위 범위를 벗어나는 관측치가 존재하지 않는 점을 고려하면 분포를 왜곡할 정도의 극단값(outlier)이 존재한다고 보기는 어렵습니다.
- 평균 하나로 이 데이터를 설명하는 것은 충분한가?
  - 이 표본에서는 설명이 가능하다고 생각합니다.
  - 중앙값과 평균의 차이가 크지 않고, 두 값 모두 사분위 범위(IQR) 내에 위치하고 있다는 점을 고려할 때, 극단값에 의한 왜곡이 거의 없으며 분포가 비교적 안정적인 구조를 가진다고 볼 수 있습니다.
  - 따라서 해당 데이터의 대표값으로 '평균'을 사용하는 것은 충분히 타당해보입니다.
  - 하지만 실제 평균은 표본의 데이터를 나타내는 하나의 지표일 뿐, 모든 표본을 대표하기에는 충분한 설명을 하기에는 무리가 있으며, 중앙값과 최빈값과 같은 여러 대표적인 지표를 복합적으로 사용하는 것이 좋습니다.

## 🔹 문제 3. 표본의 위험성 체험하기 (핵심)
- 설명

전체 데이터에서 무작위로 5명만 뽑아 평균을 계산하라.
이를 1000번 반복하고 평균의 분포를 확인하라.

```python
sample_means = []

for _ in range(1000):
    sample = np.random.choice(heights, size=5, replace=False)
    sample_means.append(sample.mean())

plt.hist(sample_means, bins=30)
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.title("Sampling Distribution (n=5)")
plt.show()

np.mean(sample_means), np.std(sample_means)
```

### 질문
- 표본 평균은 항상 같은가?
- 원래 데이터의 평균과 얼마나 차이가 나는가?
- 왜 표본 크기가 중요하다고 말하는가?

### 문제 3 : 답변

  - 표본 평균은 항상 같은가?
  - 그렇지 않습니다.
  - 표본의 데이터 값이 무엇이냐에 따라, 평균은 차이를 보이고 있습니다.
- 원래 데이터의 평균과 얼마나 차이가 나는가?
  - 대략적으로 크게는 0.254에서 작게는 0.03까지의 차이를 보였습니다.
- 왜 표본 크기가 중요하다고 말하는가?
  - 표본이 커질수록 평균은 점점 모집단 평균에 가까워집니다.(대수의 법칙) 즉, 평균의 유동성이 감소하고, 값이 안정화됩니다.
  - 작은 표본에서는 이상치 하나가 전체 결과를 왜곡하지만
큰 표본에서는 그 영향이 감소하게 됩니다. 즉, 극단값(이상치)에 대한 영향력이 미미해지게 됩니다.
  - 일반적으로 표본의 크기가 커질 수록 추정값이 더 정밀해지고, 신뢰구간이 좁아지며, 이에따라 결과에 대한 확신이 증가하게 됩니다. 즉, 표준오차가 감소하고 통계적 신뢰도가 증가하게 됩니다.

### 문제 3 : 답변


- 왜 우리는 전체가 아닌 표본으로 판단할 수밖에 없는가?
  - 우선 우리가 가진 데이터가 전체라는 보장이 없습니다. 즉, 대부분의 경우 전체 데이터를 얻는 것은 물리적으로 불가능하거나 비현실적입니다.
  - 다음으로 전체 데이터를 운이 좋게 확보한다고 해도, 시간과 비용면에서 비효율적입니다. 전체 데이터를 확보하는 것 자체가 많은 시간을 사용하는 반면, 전체에 대한 표본은 아주 극소량으로, 이 자료를 가지고, 빠른 의사결정을 내리는 편이 효율적이기 때문입니다.

- 평균 하나만 보고 결론을 내리면 위험한 이유는?
  - 평균은 데이터를 대표할 수 있는 수많은 지표중 하나일 뿐입니다. 즉, 평균은 데이터의 중심을 나타내는 대표값일 뿐이며, 분포의 구조나 변동성을 충분히 설명하지 못합니다.
  - 평균의 대표적인 문제점은 극단값에 매우 취약하다는 점입니다. 평균의 모든 값을 반영하는 특성때문에 소수의 큰/작은 값에 의해 쉽게 왜곡됩니다.
  - 또한 이러한 부분은 서로 다른 그룹의 대표적인 특성을 왜곡할 수 있습니다.

- 통계학이 “확률의 학문”이 아니라 “불확실성 관리의 학문”이라는 말의 의미는?
  - 개인적으로 통계는 '제한된 데이터로 전체를 추론하는 학문'이라고 생각합니다.
  - 이러한 점을 생각해볼때, 제한된 데이터가 전체를 대표할 수 있도록 하기위한 불확실성(확률)을 최소화하고, 의사결정에 필요한 데이터의 신뢰성이 있고, 유의미한 결론을 도출해낼 수 있도록하는 학문을 '통계학'라 생각합니다.



# LLMs 서비스 아키텍처

```python
┌───────────────────────────────┐
 │        사용자 / 직원           │
 │  (상담원, 심사자, 관리자 등)   │
 └───────────────┬───────────────┘
                 │ 질의 (업무 목적)
                 ▼
 ┌───────────────────────────────┐
 │        LLM Agent Layer         │
 │  - 질의 해석                  │
 │  - Context 범위 결정           │
 │  - Tool 호출 오케스트레이션    │
 │  ※ 실행 권한 없음             │
 └───────────────┬───────────────┘
                 │
        ┌────────┴────────┐
        ▼                 ▼
┌──────────────────┐ ┌──────────────────┐
│ 실시간 Context    │ │ 증거 검증 Context │
│ Tool (Near RT)    │ │ Tool (DW 기반)   │
│                  │ │                  │
│ - OGG 복제 데이터 │ │ - 이력 / 패턴     │
│ - 분리된 Cluster │ │ - 집계 / 요약     │
│ - 범위 제한 조회 │ │ - 설명 가능성     │
└────────┬─────────┘ └────────┬─────────┘
         │                       │
         ▼                       ▼
 ┌─────────────────────────────────────┐
 │        Evidence-based Reasoning      │
 │  - 실시간 vs 이력 정합성 비교        │
 │  - 충돌 / 불일치 탐지               │
 │  - Confidence Score 산출            │
 └───────────────┬─────────────────────┘
                 │
                 ▼
 ┌─────────────────────────────────────┐
 │         응답 + 신뢰도 표시           │
 │  - 설명 / 요약 / 추천               │
 │  - Confidence Score                 │
 │  ※ 최종 결정은 사람                │
 └─────────────────────────────────────┘
```
